```python
import numpy as np
from scipy import stats

original = np.array([24.0, 30.5, 35.2, 36.8, 37.6, 39.8, 40.8, 38.5, 37.8, 43.0])
modified = np.array([26.8, 29.5, 31.8, 36.0, 39.1, 40.8, 40.2, 40.6, 39.6, 41.2])

mean_original = np.mean(original)
mean_modified = np.mean(modified)

t_stat, p_val = stats.ttest_rel(modified, original)

print(f"Original Mean: {mean_original:.4f}")
print(f"Modified Mean: {mean_modified:.4f}")
mean_diff = np.mean(modified - original)
print(f"Mean Difference (Modified - Original): {mean_diff:.4f}")
print(f"T-statistic: {t_stat:.4f} | P-value: {p_val:.4f}")
```
